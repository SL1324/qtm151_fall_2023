(d) Create a function with four inputs $f(y,x,b0,b1)$

<font size = "3">

$\qquad e2 = (y - b_0 - b_1 * x)^2$

- The function should return the mean of e2
- This is the mean-square error (mse)
- Call the function "fn_mse"
- Let $y$ be the column "Potential"
- Let $x$ be the column "Age" 
- Does <br>
$\qquad (b_0 = 80, b_1 = 0.4)$ or <br>
$\qquad (b_0 = 80, b_1 =-0.3)$ or <br>
$\qquad (b_0 = 80, b_1 = -0.5)$ <br>
produce the lowest value of mse?
- Print your results

````python
# Write your own code
def fn_mse(y, x, b0, b1):
    e2 = ((y - b0 - b1 * x) ** 2).mean()
    return e2


parameters = [(80, 0.4), (80, -0.3), (80, -0.5)]

mse_values = {}
mse_values['(b0=80, b1=0.4)'] = fn_mse(80, 0.4, dt['Potential'], dt['Age'])
mse_values['(b0=80, b1=-0.3)'] = fn_mse(80, -0.3, dt['Potential'], dt['Age'])
mse_values['(b0=80, b1=-0.5)'] = fn_mse(80, -0.5, dt['Potential'], dt['Age'])
for key, value in mse_values.items():
    print(f"For {key}, MSE = {value:.2f}")

lowest= min(mse_values, key=mse_values.get)
print(f"{lowest} produces the lowest value of MSE.")

   
````

(e) Split a dataset into subsets

<font size = "3">

- Create ``` list_status = [0,1]```
- Create ``` list_prob = [0.5, 0.5]```
- Create a new variable $n$ with the number of rows of the dataset
- Define a random seed that equals 42

- Use 
```python
    np.random.choice(list_status,
                     size = n,
                     p = list_prob)
```
$\qquad$ to create a new column called "status" <br>
$\qquad$ which randomly assigns each row to "0" or "1" <br>
$\qquad$ with equal probability
- Use .query() to subset all observations with status == 0.<br>
store it in a new dataset called "data0"
- Use .query() to subset all observations with status == 1.<br>
store it in a new dataset called "data1"



````python
# Write your own code
list_status = [0,1]
list_prob = [0.5, 0.5]
np.random.seed(42)
n = len(dt)
dt['status'] = np.random.choice(list_status, size=n, p=list_prob)
data0 = dt.query('status== 0')
data1 = dt.query('status == 1')
data0
````

(f) Create two overlapping histogram plots

<font size = "3">

- This question requires the two datasets you created in (e)
- Display the distribution of "Age" for "data0" and "data1"
- Use the "alpha" option to make the graphs semitransparent
- You will need to add a legend, label the axes, and the title

````python
# Write your own code
plt.figure(figsize=(10, 6))
plt.hist(data0['Age'], bins=15, alpha=0.5, label='Data0', color='purple')
plt.hist(data1['Age'], bins=15, alpha=0.5, label='Data1', color='blue')
plt.title('Distribution of Age for Data0 and Data1')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
# "The goal of this graph is to illustrate that random assignment results in very similar distributions between two groups, "Status 0" and "Status 1." In the plot, the histograms for the "Age" distribution for both groups are shown. The semi-transparent bars show that the distributions overlap significantly. Therefore, there are no big differences in the "Age" distribution between the two  groups. This suggests that random assignment was effective in creating groups with similar distributions."
````
