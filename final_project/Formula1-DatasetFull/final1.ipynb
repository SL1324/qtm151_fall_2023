```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib as mpl # visualizations
import matplotlib.pyplot as plt
```

```python
results=pd.read_csv('raw_data/results.csv') #import results dataset
races=pd.read_csv("raw_data/races.csv")
drivers=pd.read_csv("raw_data/drivers.csv")
pit_stops=pd.read_csv("raw_data/pit_stops.csv")
```

```python
results
```

```python
subset=results.query("milliseconds.isnumeric()==False")
list_unique=pd.unique(subset["milliseconds"])
print(list_unique)
```

```python
list_old=['\\N']
list_new=[0]
results["milliseconds"]=results["milliseconds"].replace(list_old,list_new)
results["milliseconds_numeric"] = pd.to_numeric(results["milliseconds"])
results=results.query('milliseconds_numeric >0')
results.to_csv("raw_data/results.csv")
results.dtypes
```

```python
pit_stops = pit_stops.rename(columns={'milliseconds': 'stopinmilliseconds'})
print(pit_stops.dtypes)
display(pit_stops)
sum_stoptimes=pit_stops.groupby(["raceId","driverId"]).agg(sum_milliseconds=('stopinmilliseconds','sum'))
display(sum_stoptimes)
```

```python
results
```

```python
pit_races=pd.merge(pit_stops[['raceId','stopinmilliseconds']],
                               races[['raceId','year','date']],
                               on='raceId',
                           how="right")
final_results=pd.merge(pit_races[["raceId","stopinmilliseconds",'year','date']],
                       results[['raceId','driverId','milliseconds_numeric']],
                           on="raceId",
                           how="right")
```

```python
season_results=(final_results.query("year==2012"))
season_results['date_new'] = pd.to_datetime(season_results["date"], format = "%Y-%m-%d")
display(season_results)
race_results_2012=(season_results.groupby(["raceId","driverId"]).agg(milliseconds_numeric=('milliseconds_numeric','mean'),))
final_re=pd.merge(race_results_2012,sum_stoptimes,
                           on="raceId",
                           how="right")
display(final_re)


```

```python
race_results_2012_sorted=race_results_2012.sort_values(by='date_new')
display(race_results_2012_sorted)
race_re2012=race_results_2012_sorted.apply(lambda x: x.sort_values(by='milliseconds_numeric')).reset_index(drop=True)
display(race_re2012)
```

```python
conditional_average = season_results[season_results['raceId'] ==860].groupby("driverId")['stopinmilliseconds'].sum()
display(conditional_average)
```
